```python
#create missing values with NaN

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score


dataset = pd.read_csv('pima-indians-diabetes.csv',header=None)
print(dataset.head(10))
data = dataset.values
dataset.columns = [0,1,2,3,4,5,6,7,8]

print('Number of instances = %d' % (dataset.shape[0]))
print('Number of attributes = %d' % (dataset.shape[1]))
dataset[[1,2,3,4,5]] = dataset[[1,2,3,4,5]].replace(0, np.NaN)
print(dataset.head(10))

print('Number of missing values:')
for col in dataset.columns:
    print('\t%s: %d' % (col,dataset[col].isnull().sum()))

X_train, X_test, y_train, y_test = train_test_split(data[:,:-1], data[:,8], test_size=0.3, random_state=0)

clf_DT = tree.DecisionTreeClassifier()
clf_DT = clf_DT.fit(X_train, y_train)

hasil_testing_DT = clf_DT.predict(X_test)
accuracy_DT = accuracy_score(y_test, hasil_testing_DT)
print('Accuracy =',accuracy_DT)
```

```python
#drop Missing values (NaN) menggunakan fungsi dropna

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

dataset_drop = pd.read_csv('pima-indians-diabetes.csv',header=None)
dataset_drop.columns = [0,1,2,3,4,5,6,7,8]
dataset_drop[[1,2,3,4,5]] = dataset_drop[[1,2,3,4,5]].replace(0, np.NaN)

dataset_drop.dropna(inplace=True)
print(dataset_drop.shape)
print(dataset_drop.head(10))
print('Number of missing values:')
for col in dataset_drop.columns:
    print('\t%s: %d' % (col,dataset_drop[col].isnull().sum()))

data_drop = dataset_drop.values
X_train, X_test, y_train, y_test = train_test_split(data_drop[:,:-1], data_drop[:,8], test_size=0.3, random_state=0)

clf_DT = tree.DecisionTreeClassifier()
clf_DT = clf_DT.fit(X_train, y_train)

hasil_testing_DT = clf_DT.predict(X_test)
accuracy_DT = accuracy_score(y_test, hasil_testing_DT)
print('Accuracy drop NaN=',accuracy_DT)
```

```python
#imputer mean menggunakan numpy

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

dataset_mean= pd.read_csv('pima-indians-diabetes.csv',header=None)
dataset_mean.columns = [0,1,2,3,4,5,6,7,8]
dataset_mean[[1,2,3,4,5]] = dataset_mean[[1,2,3,4,5]].replace(0, np.NaN)

dataset_mean = dataset_mean.fillna(dataset_mean.median())
print(dataset_mean.shape)
print(dataset_mean.head(10))
print('Number of missing values:')
for col in dataset_mean.columns:
    print('\t%s: %d' % (col,dataset_mean[col].isnull().sum()))

data_mean = dataset_mean.values
X_train, X_test, y_train, y_test = train_test_split(data_mean[:,:-1], data_mean[:,8], test_size=0.3, random_state=0)

clf_DT = tree.DecisionTreeClassifier()
clf_DT = clf_DT.fit(X_train, y_train)

hasil_testing_DT = clf_DT.predict(X_test)
accuracy_DT = accuracy_score(y_test, hasil_testing_DT)
print('Accuracy imputer mean=',accuracy_DT)
```

```python
#imputer with mean using sklearn SimpleImputer

import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

dataset = pd.read_csv('pima-indians-diabetes.csv', header=None)

dataset[[1,2,3,4,5]] = dataset[[1,2,3,4,5]].replace(0, np.NaN)

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(dataset)
dataset_mean = imp_mean.transform(dataset)
df = pd.DataFrame(dataset_mean)
print(df[:10])

X_train, X_test, y_train, y_test = train_test_split(dataset_mean[:,:-1], dataset_mean[:,8], test_size=0.3, random_state=0)

clf_DT = tree.DecisionTreeClassifier()
clf_DT = clf_DT.fit(X_train, y_train)

hasil_testing_DT = clf_DT.predict(X_test)
accuracy_DT = accuracy_score(y_test, hasil_testing_DT)
print('Accuracy mean imputer=',accuracy_DT)
```

```python
#imputer with most frequent using sklearn SimpleImputer

import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

dataset = pd.read_csv('pima-indians-diabetes.csv', header=None)

dataset[[1,2,3,4,5]] = dataset[[1,2,3,4,5]].replace(0, np.NaN)

imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mode.fit(dataset)
dataset_mode = imp_mode.transform(dataset)
df = pd.DataFrame(dataset_mode)
print(df[:10])

X_train, X_test, y_train, y_test = train_test_split(dataset_mode[:,:-1], dataset_mode[:,8], test_size=0.3, random_state=0)

clf_DT = tree.DecisionTreeClassifier()
clf_DT = clf_DT.fit(X_train, y_train)

hasil_testing_DT = clf_DT.predict(X_test)
accuracy_DT = accuracy_score(y_test, hasil_testing_DT)
print('Accuracy modus imputer=',accuracy_DT)
```

```python
#imputer menggunakan impyute https://impyute.readthedocs.io/en/latest/

import impyute as impy

dataset = pd.read_csv('pima-indians-diabetes.csv', header=None)

dataset[[1,2,3,4,5]] = dataset[[1,2,3,4,5]].replace(0, np.NaN)

imp_mean = impy.mean(dataset)
print(imp_mean[:10])

imp_median = impy.median(dataset)
print(imp_median[:10])

imp_mode = impy.mode(dataset)
print(imp_mode[:10])

imp_random = impy.random(dataset)
print(imp_random[:10])

imp_em = impy.em(dataset)
print(imp_em[:10])

imp_knn = impy.fast_knn(dataset, k=15)
print(imp_knn[:10])
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
import impyute as impy
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

dataset = pd.read_csv('pima-indians-diabetes.csv', header=None)

dataset[[1,2,3,4,5]] = dataset[[1,2,3,4,5]].replace(0, np.NaN)

imp_em = impy.em(dataset)
data_em = imp_em.values

X_train, X_test, y_train, y_test = train_test_split(data_em[:,:-1], data_em[:,8], test_size=0.3, random_state=0)

clf_DT = tree.DecisionTreeClassifier()
clf_DT = clf_DT.fit(X_train, y_train)

hasil_testing_DT = clf_DT.predict(X_test)
accuracy_DT = accuracy_score(y_test, hasil_testing_DT)
print('Accuracy EM imputer=',accuracy_DT)

imp_knn = impy.fast_knn(dataset, k=15)
data_knn = imp_knn.values

X_train, X_test, y_train, y_test = train_test_split(data_knn[:,:-1], data_knn[:,8], test_size=0.3, random_state=0)

clf_DT = tree.DecisionTreeClassifier()
clf_DT = clf_DT.fit(X_train, y_train)

hasil_testing_DT = clf_DT.predict(X_test)
accuracy_DT = accuracy_score(y_test, hasil_testing_DT)
print('Accuracy kNN k=15 imputer=',accuracy_DT)
```
